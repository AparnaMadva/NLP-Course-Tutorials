#### Tokenization

```python
# Sentence tokenization
# Split a paragraph into sentences
from nltk.tokenize import sent_tokenize

sample_paragraph = """Given a set of rules for conversation, humans can interact with a chatbot, rate its replies for compliance with the rules,and discover failure cases.
Classifiers trained on data generated through such interactions can tell the bot when it has broken a rule.
Then it can learn to generate output that conforms with the rules."""

tokenized_paragraph = sent_tokenize(sample_paragraph)

print("No of sentences: ", len(tokenized_paragraph),"\n")
tokenized_paragraph
```

```python
# Word tokenization
# Split a sentence into word tokens
from nltk.tokenize import word_tokenize

sample_sentence = """Sparrow is not Google’s answer to ChatGPT, it preceded OpenAI’s offering by several weeks"""

tokenized_sentence = word_tokenize(sample_sentence)

print("No of words: ",len(tokenized_sentence),"\n")
tokenized_sentence
```

#### Stopword removal

```python
# stopword removal
from nltk.corpus import stopwords

en_stopwords = set(stopwords.words('english'))

print("Number of stopwords: ",len(en_stopwords))
en_stopwords
```

```python
# stopword list is also available for below languages
stopwords.fileids()
```

```python
# hinglish stopwords
stopwords.words('hinglish')
```

```python
sample_text = """Beginning in November 2022 or earlier, CNET’s editors used an unnamed,proprietary model built by its parent company Red Ventures to produce articles on personal finance.The editors, who either published the model’s output in full or wove excerpts into material written by humans,were responsible for ensuring the results were factual. Nonetheless, they published numerous errors and instances of possible plagiarism."""

# first tokenize the sample text
tokenized_text = word_tokenize(sample_text)

# remove the stopwords from the sample text
filtered_text = [token for token in tokenized_text if token not in en_stopwords]

print("Original text:\n",sample_text)
print("After stopword removal:\n"," ".join(filtered_text))
```

#### Stemming

```python
# Stemming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

sample_sentence = """Sparrow is not Google’s answer to ChatGPT, it preceded OpenAI’s offering by several weeks"""
tokenized_sentence = word_tokenize(sample_sentence) #tokenize the sentence

stemmed_sentence = [stemmer.stem(token) for token in tokenized_sentence] #stemming the tokens

print("Original tokens:\n",tokenized_sentence,"\n")
print("After Porter stemmer:\n",stemmed_sentence)
```

```python
sample_words = ['run', 'runs', 'ran', 'running', 'runner']

for w in sample_words:
    print(w, " \t: ", stemmer.stem(w))
```

#### Lemmatization

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

tokenized_sentence = word_tokenize("Sparrow is not Google’s answer to ChatGPT, it preceded OpenAI’s offering by several weeks")  # tokenize sentence before lemmatizing

lemmas_sentence = [lemmatizer.lemmatize(token) for token in tokenized_sentence]

print("Original tokens:\n",tokenized_sentence,"\n")
print("After Lemmatizing:\n",lemmas_sentence)
```

```python
sample_words = ['run', 'runs', 'ran', 'running', 'runner']

for w in sample_words:
    print(w, " \t: ", lemmatizer.lemmatize(w))
```

```python
lemmatizer.lemmatize('running', pos='v')
```

```python
from nltk.corpus import brown

# brown Corpus is a convenient resource for studying systematic differences between genres
brown.categories()
```

```python
brown_text = brown.words(categories='government')

print(len(brown_text))
brown_text
```

```python
brown_sentences = ' '.join(brown_text)
brown_sentences
```

```python
# generate wordcloud

#wordcloud = WordCloud().generate(brown_sentences)
wordcloud = WordCloud(background_color="white", max_font_size=100).generate(brown_sentences)
plt.figure(figsize=(10,8))
# display the wordcloud, 'bilinear' for interpolation is to make the image appear smooth
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
len(stopwords)
```

```python
# wordcloud after removing stopwords
wordcloud=WordCloud(stopwords=stopwords, background_color='white', max_font_size=100).generate(brown_sentences)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
stopwords.update(['will','year','made','may'])

# wordcloud after adding few other stopwords
wordcloud=WordCloud(stopwords=stopwords, background_color='white', max_font_size=100).generate(brown_sentences)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# number of words can also be limited for the wordcloud
wordcloud=WordCloud(stopwords=stopwords, background_color='white', max_font_size=100, max_words=100).generate(brown_sentences)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

#### textrank implementation in python

```python
# TextRank is a graph based ranking model for text processing useful for keyword and sentnece extraction
!pip install pytextrank

import spacy  # load the spacy model
nlp = spacy.load("en_core_web_sm")
```

```python
# more about PyTextRank, tutorials at https://derwen.ai/docs/ptr/
# examples in github, https://github.com/DerwenAI/pytextrank/tree/main/examples
import pytextrank

# add PyTextRank to the spacy pipeline
nlp.add_pipe("textrank")
```

```python
nlp.pipe_names
```

```python
text ="""As we enter the new year, lets view 2023 not as a single year, but as the first of more in which we will accomplish our long-term goals.
Some results take a long time to achieve, and even though we may take actions that bring those results closer,
we can do it more effectively if we envision a path rather than simply going from milestone to milestone."""

doc = nlp(text)
# examine the top-ranked phrases in the document
for phrase in doc._.phrases:
    print(phrase)
```

```python
tr = doc._.textrank
lemma_graph = tr.lemma_graph  #returns a networkx digraph object, graph constructed by iterating through sentences

tr.edge_list[:5]  #list of weighted edges for the lemma graph
```

```python
# summarizing a paragraph using textrank

text = """Humans can select the right pieces of knowledge and paste them together to form a relevant explanation, answer, or plan.
Moreover, given a set of variables, humans are pretty good at deciding which is a cause of which.
Current AI techniques don’t come close to this human ability to generate reasoning paths. Often, they’re highly confident that their decision is right, even when it’s wrong.
Such issues can be amusing in a text generator, but they can be life-threatening in a self-driving car or medical diagnosis system.
Current systems behave in these ways partly because they’ve been designed that way.
For instance, text generators are trained simply to predict the next word rather than to build an internal data structure
that accounts for the concepts they manipulate and how they are related to each other.
I think we can design systems that track the meanings at play and reason over them while keeping the numerous advantages of current deep learning methodologies.
In doing so, we can address a variety of challenges from excessive sample complexity to overconfident incorrectness."""


nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("textrank")
doc = nlp(text)
tr = doc._.textrank
```

```python
for sent in tr.summary(limit_phrases=15, limit_sentences=2):
#for sent in tr.summary(limit_phrases=10):
    print(sent)
```
